```
# 개념정리

# 텐서(tensor)
# 플레이스홀더(placeholder)
# 변수(variable)
```

```
import tensorflow as tf
print(tf.__version__)
```

```
# tf.constant로 상수를 hello 변수에 저장

hello = tf.constant('Hello,TensorFlow!')
print(hello)                    
```

```
# 랭크(Rank)
  # 랭크가 0이면 스칼라, 1이면 벡터, 2면 행렬 3 이상이면 n-Tensor 또는 n차원 텐서
# 셰이프(Shape)
  # 각 차원의 요소 개수로, 텐서의 구조를 설명
```

```
a = tf.constant(10)
b = tf.constant(32)
c = tf.add(a,b)
print(c)
```

```
# 텐서플로 프로그램의 구조가 다음의 두 가지로 분리되어 있음
  # 1. 그래프 생성
  # 2. 그래프 실행
  
# 그래프: 텐서들의 연산 모음
# 지연실행: 텐서와 텐서의 연산들을 먼저 정의하여 그래프를 만들고, 이후 필요할 때 연산을 실행하는 코드를 넣어 실행
```

```
# 그래프의 실행은 Session 안에서 이뤄져야 하며, 다음과 같이 Session 객체와 run 메서드를 이용

sess = tf.Session()

print(sess.run(hello))
print(sess.run([a,b,c]))

sess.close()
```

```
# 3.2 플레이스홀더와 변수
# 플레이스 홀더: 그래프에 사용할 입력값을 나중에 받기 위해 사용하는 매개변수
# 변수: 그래프를 최적화 하는 용도로 텐서플로가 학습한 결과를 갱신하기 위해 사용하는 변수
```

```
# 플레이스 홀더는 다음과 같이 사용

# None은 크기가 정해지지 않았음을 의미합니다.
X = tf.placeholder(tf.float32,[None,3])
print(X)
```

```
# Placeholder라는 (?, 3) 모양의 float32 자료형을 가진 텐서가 생성됨
```

```
x_data = [[1,2,3],[4,5,6]]
```

```
# 다음은 W,b 변수들을 정의
# W는 [3,2] 행렬 형태
# b는 [2,1] 행렬 형태의 텐서로 tf.random_normal 함수를 이용해 정규분포의 무작위 값으로 초기화

W = tf.Variable(tf.random_normal([3,2]))
b = tf.Variable(tf.random_normal([2,1]))
```

```
# 물론 이와 같이 직접 원하는 텐서의 형태의 데이터를 만들어서 넣어 줄 수도 있음.

W = tf.Variable([[0.1,0.1], [0.2, 0.2], [0.3, 0.3]])
```

```
# 다음으로 입력값과 변수들을 계산할 수식을 작성
# W와 X가 행렬이기 떄문에 tf.matmul 함수를 사용해야 함.
# 행렬이 아닌 경우에는 단순히 곱셈 연산자(*)나 tf.mul 함수를 사용

expr = tf.matmul(X,W) + b
```

```
# 행렬곱
  # 행렬곱 A x B에 대하여, 행렬 A의 열 수와 행렬 B의 행수는 같아야 함
  # 행렬곱 A x B를 계산한 행렬 AB의 크기는 A의 행 개수와 B의 열개수가 된다.
  
```

```
# 설정한 텐서들과 계산된 그래프의 결과를 확인해 보겠습니다.

sess = tf.Session()

# 앞에 정의한 변수들을 초기화
sess.run(tf.global_variables_initializer()) 

print("=== x_data===")
print(x_data)
print("=== W ===")
print(sess.run(W))
print("=== b ===")
print(sess.run(b))
print("=== expr ===")

# feed_dict: 그래프를 실행 할 떄 사용할 입력값 지정
print(sess.run(expr, feed_dict={X: x_data})) 

sess.close()
```

```
# expr 수식에는 X, W, b를 사용했는데, 이중 X가 플레이스홀더라 X에 값을 넣어주지 않으면
# 계산에 사용할 값이 없으므로 에러
# 따라서 미리 정의해둔 x_data를 X 값으로 넣음.
```

```
# 3.3 선형 회귀 모델 구현하기
  # 선형 회귀: 주어진 x와 y값을 가지고 서로간의 관계를 파악하는 것
```

```
# 텐서플로의 최적화 함수를 이용해 X와 Y의 상관관계를 분석하는 기초적인 선형 회귀 모델

x_data = [1,2,3]
y_data = [1,2,3]
```

```
# x와 y의 상관관계를 설명하기 위한 변수들인 W와 b를 각각 -1.0부터 0.1 사이의 균등분포를 가진 무작위값으로 초기화

W = tf.Variable(tf.random_uniform([1], -1.0,1.0))
b = tf.Variable(tf.random_uniform([1], -1.0,1.0))
```

```
# 자료를 입력 받을 플레이스 홀더 설정

X = tf.placeholder(tf.float32, name="X")
Y = tf.placeholder(tf.float32, name="Y")
```

```
# X와 Y의 상관관계(선형관계)를 분석하기 위한 수식 작성
# W와의 곱과 b와의 합을 통해 X와 Y의 관계를 설명하겠다.
# 즉, X가 주어졌을 때 Y를 만들어 낼 수 있는 W와 b를 찾아내겠다.
# W 가중치, b는 편향

hypothesis = W * X + b
```

```
# 손실함수 작성
# 손실함수: 한 쌍(x, y)의 데이터에 대한 손실값을 계산하는 함수
# 손실값: 실제값과 모델로 예측한 값이 얼마나 차이가 나는가를 나타내는 값

# 즉, 손실값이 작을수록 그 모델이 X와 Y의 관계를 잘 설명하고 있다는 뜻
# 주어진 X 값에 대한 Y값을 정확하게 예측 할 수 있다라는 뜻

# 비용: 이 손실을 전체 데이터에 대해 구한 경우
# 학습: 변수들의 값을 다양하게 넣어 계산해보면서 이 손실값을 최소화하는 W와 b의 값을 구하는 것
# 손실 값으로 '예측값과 실제값의 거리'를 가장 많이 사용함. 우리도 이를 사용
# 즉, 손실값은 예측값에서 실제값을 뺸 뒤 제곱하여, 그리고 비용은 모든 데이터에 대한 손실값의 평균을 내어 구함

cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```
# 경사하강법 최적화 함수를 이용해 손실값을 최소화하는 연산 그래프를 생성
# 경사하강법: 최적화 방법 중 가장 기본적인 알고리즘, 함수의 기울기를 구하고 기울기가 낮은 쪽으로 계속 이동
# 시키면서 최적의 값을 찾아 나가는 방법

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.1)
train_op = optimizer.minimize(cost)
```

```
# 최적화 함수: 가중치와 편향 값을 변경해가면서 손실값을 최소화하는 가장 최적화된 가중치와 편향 값을
# 찾아주는 함수
# learning_rate: 최적화 함수의 매개변수, 학습률은 얼마나 '급하게' 할것인가를 설정하는 것
# 값이 너무 크면 최적의 손실값을 찾지 못하고 지나치게 되고
# 값이 너무 작으면 학습 속도가 매우 느려집니다.
# 하이퍼파라미터 : 학습을 진행하는 과정에 영향을 주는 변수
# 이를 잘 튜닝하는 것이 큰 과제
```

```
# 이제 선형 회귀 모델을 다 만들었으니, 그래프를 실행해 학습 시키고 결과 확인
# 세션을 생성하고 변수들을 초기화
# 파이썬 with 기능을 이용해 세션 블록을 만들고 세션 종료를 자동으로 처리

with tf.Session() as sess:
  sess.run(tf.global_variables_initializer())

# 다음은 최적화를 수행하는 그래프인 train_op를 실행
# 실행 시마다 변화하는 손실값을 출력
# 학습은 100번 수행, feed_dict 매개변수 통해, 상관관계를 알아내고자 하는 데이터인 x_data와 y_data입력

  for step in range(100):
    _, cost_val = sess.run([train_op,cost], feed_dict={X: x_data, Y: y_data})
    
    print(step, cost_val, sess.run(W), sess.run(b))
```

```
# 위와 같이 손실값이 점점 줄어든다면 학습이 정상적으로 이뤄지고 있는것
```

```
# 학습에 사용되지 않았던 값인 5와 2.5를 X값으로 넣고 결과를 확인

with tf.Session() as sess:
  sess.run(tf.global_variables_initializer())

  for step in range(100):
    _, cost_val = sess.run([train_op,cost], feed_dict={X: x_data, Y: y_data})
    
    print(step, cost_val, sess.run(W), sess.run(b))
    
  print("\n=== Test ===")
  print("X:5, Y:", sess.run(hypothesis, feed_dict={X: 5}))
  print("X:2.5, Y:", sess.run(hypothesis, feed_dict={X: 2.5}))
```

```
# 기대한 대로 X가 5일 때는 Y값으로 5를, 2.5 일때는 2.5를 정확히 예측해 냈습니다.
```
